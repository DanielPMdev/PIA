# 3.1 Crea un DataFrame a partir del siguiente diccionario de Series:
#     personas = {
#         'peso': pd.Series([84, 90, 56, 64], ['Santiago', 'Pedro', 'Ana', 'Julia']),
#         'altura': pd.Series({'Santiago': 187, 'Pedro': 178, 'Julia': 170, 'Ana': 165}),
#         'hijos': pd.Series([2, 3], ['Pedro', 'Julia'])
#     }

```python
personas = {
    'peso': pd.Series([84, 90, 56, 64], ['Santiago', 'Pedro', 'Ana', 'Julia']),
    'altura': pd.Series({'Santiago': 187, 'Pedro': 178, 'Julia': 170, 'Ana': 165}),
    'hijos': pd.Series([2, 3], ['Pedro', 'Julia'])
    }

df = pd.DataFrame(personas)
df
```


# 3.2 Observa cómo se rellenan los valores faltantes con NaN.

```python
df.isna().sum()
```

# 3.3 Crea un DataFrame nuevo con las columnas ['altura', 'peso'] y el índice ['Ana', 'Julia', 'Santiago'].

```python
df2 = df[['altura', 'peso']]
df2.index = ['Ana', 'Julia', 'Pedro', 'Santiago']
df2
```

# 3.4 Crea un DataFrame a partir de una lista de listas. Especifica columnas e índices.

```python
datos = [
    [84, 187, 2],
    [90, 178, 3],
    [56, 165, 0],
    [64, 170, 1]
]

df3 = pd.DataFrame(datos, columns=['peso', 'altura', 'hijos'], index=['Santiago', 'Pedro', 'Ana', 'Julia'])
df3
```

```python
max_altura = 170
df.eval("altura > @max_altura", engine="python")
```
