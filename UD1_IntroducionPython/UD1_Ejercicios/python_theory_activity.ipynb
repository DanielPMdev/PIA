# Herencia

En los lenguajes de programación la herencia tiene una cierta similitud con los seres humanos y la biología en general. Nosotros, seguro que compartimos similitudes y rasgos con nuestros padres, abuelos y familiares, características que hemos heredado de ellos: el carácter, color del pelo, etc. En Python ocurre algo similar. Vamos a definir la clase persona con unos cuantas propiedades.

```python
class Persona:
  def __init__(self, nombre: str, apellido: str, edad: int):
    self.nombre = nombre
    self.apellido = apellido
    self.edad = edad
```

Ahora definiremos la clase estudiante. Un estudiante también es una persona pero quizás con unas propiedades extra como la universidad a la que pertenece.

```python
class Estudiante(Persona):
  def __init__(self, nombre: str, apellido: str, edad: int, universidad: str):
    self.universidad = universidad
    super().__init__(
        nombre=nombre,
        apellido=apellido,
        edad=edad
    )
```

En el constructor de la clase estudiante primero hemos inicializado el atributo universidad y posteriormente, hemos aprovechado el constructor de la case Persona para terminar de inicializar el objeto. Ahora, nuestro estudiante, también tiene las propiedades de la clase Persona: nombre, apellido y edad.

Además de los atributos, también se pueden heredar métodos. Vamos a expandir la clase Persona.

```python
class Persona:
  def __init__(self, nombre: str, apellido: str, edad: int):
    self.nombre = nombre
    self.apellido = apellido
    self.edad = edad

  def imprime_nombre(self):
    print(self.nombre, self.apellido)

class Estudiante(Persona):
  def __init__(self, nombre: str, apellido: str, edad: int, universidad: str):
    self.universidad = universidad
    super().__init__(
        nombre=nombre,
        apellido=apellido,
        edad=edad
    )
```

Vamos a ver un ejemplo de como podemos acceder al método imprime_nombre desde un objeto de la clase estudiante.

```python
estudiante1 = Estudiante(
    nombre='Lucas',
    apellido='Hernandez',
    edad=25,
    universidad='UPV'
)

estudiante1.imprime_nombre()
```

También podemos sobreescribir métodos definidos en la clase padre. A continuación definiremos un método dentro de la clase Estudiante que sustituirá al definido en la clase Persona. Cualquier clase que herede de la clase Persona seguirá disponiendo del método imprime_nombre original.

```python
class Estudiante(Persona):
  def __init__(self, nombre: str, apellido: str, edad: int, universidad: str):
    self.universidad = universidad
    super().__init__(
        nombre=nombre,
        apellido=apellido,
        edad=edad
    )

  def imprime_nombre(self):
    print(self.nombre, self.apellido, self.universidad)

estudiante1 = Estudiante(
    nombre='Lucas',
    apellido='Hernandez',
    edad=25,
    universidad='UPV'
)

estudiante1.imprime_nombre()
```

Existe la posibilidad de definir métodos en la clase padre que incluyen métodos que únicamente se han definido en la clase hijo. Por ejemplo, vamos a definir la clase Trabajador que hereda de la clase Persona.

```python
class Persona:
  def __init__(self, nombre: str, apellido: str, edad: int):
    self.nombre = nombre
    self.apellido = apellido
    self.edad = edad

  def imprime_nombre(self):
    print(self.nombre, self.apellido)

  def imprime_ocupacion(self):
    print(self.obten_actividad())

class Trabajador(Persona):
  def __init__(self, nombre: str, apellido: str, edad: int, empresa: str):
    self.empresa = empresa
    super().__init__(
        nombre=nombre,
        apellido=apellido,
        edad=edad
    )

  def imprime_nombre(self):
    print(self.nombre, self.apellido, self.empresa)

  def obten_actividad(self):
    return 'Trabajador'
```

En la clase Persona hemos definido el método imprime_ocupacion(), dentro de este metodo accedemos a obten_actividad() el cual no está definido dentro de la clase Persona pero sí en la clase Trabajador. Si intentamos acceder a dicho método desde la clase Persona obtendremos un error, pero no si lo hacemos desde una instancia de un objeto trabajador.

```python
persona1 = Persona(
    nombre='Juan',
    apellido='Garcia',
    edad=21
)

#persona1.imprime_ocupacion()
```

```python
trabajador1 = Trabajador(
    nombre='Luis',
    apellido='Sanchez',
    edad=31,
    empresa='IES.Abastos'
)

trabajador1.obten_actividad()
```

A continuación vamos a descubrir algunas de las particularidades de Python y que permiten enriquecer nuestro código y añadirle complejidad.

Cómo hemos visto hasta el momento, todos los métodos que hemos ido definiendo han estado asociados a la instancia y no a la clase. Es decir, que para poder acceder al método imprime_nombre(), primero hemos de crear un objeto del tipo Persona. Sin embargo, existen métodos que permiten ir asociados a la clase Persona y no a la instancia.

Imaginemos que tenemos la clase Vehiculo, y dentro de ella tenemos un método que permite pasar de kilómetros por hora a millas por hora. Este método tiene relación directa con el vehículo pero no necesita información asociada al objeto, por lo que se puede acceder al mismo sin necesidad de instanciar la clase.

```python
class Vehiculo:
  def __init__(self, marca: str, modelo: str, potencia: int):
    self.marca = marca
    self.modelo = modelo
    self.potencia = potencia

  @staticmethod
  def transforma_kmh_a_mph(velocidad_kmh: int):
    return velocidad_kmh*0.6214
```

Hemos definido el método transforma_kmh_a_mph() y debemos observar lo siguiente: no hemos incluido el atributo self a la hora de definir los parámetros en la definición del método y además, hemos añadido la expresión @staticmethod antes de crear el método. De esta forma, podemos crear métodos que estén relacionados con la clase pero que se puedan acceder sin necesidad de instanciar la clase. Vamos a comprobar como podemos acceder a ellos.

```python
Vehiculo.transforma_kmh_a_mph(velocidad_kmh=100)
```

La expresion @staticmethod permite de una forma muy simplificada poder definir este tipo de métodos gracias al uso de decoradores, algo que veremos a continuación.

En ocasiones, nos encontramos con la tarea de definir clases que básicamente son estructuras que almacenan datos en diferentes atributos. La tarea de definir dichas clases se puede simplificar con el uso de otro decorador, en este caso @dataclass.

Cuando definimos este tipo de clases junto al decorador, no tenemos ni que definir un constructor! Vamos a ver un ejemplo.

```python
class Persona:
  def __init__(self, nombre: str, apellido: str, edad: int):
    self.nombre = nombre
    self.apellido = apellido
    self.edad = int
```

A continuación vamos a ver como se puede simplificar la tarea con el decordador @dataclass.

```python
from dataclasses import dataclass

@dataclass
class Persona:
  nombre: str
  apellido: str
  edad: int
```

Para instanciar un objeto no tenemos que hacer nada diferente. Su uso es igual al de una clase definida de la forma tradicional. Podemos definir métodos dentro de un @dataclass tal y cómo hacíamos anteriormente.

```python
from dataclasses import dataclass

@dataclass
class Persona:
  nombre: str
  apellido: str
  edad: int

  def imprime_nombre(self):
    print(self.nombre, self.apellido)

persona1 = Persona(
    nombre='Pedro',
    apellido='Lopez',
    edad=25
)

persona1.imprime_nombre()
```

Además de los decoradores @staticmethod, y @dataclass, existen otros dos decoradores más que son importantes y nos facilitan mucho nuestra labor.

El primero de ellos es @property. Imaginemos que tenemos una clase PeriodoTemporal cuyos atributos principales son fecha_inicio y fecha_fin. Imagina que queremos obtener un array con todas las fechas comprendidas entre fecha_inicio y fecha_fin. ¿Tendría sentido almacenar toda esta información en la clase? La respuesta no, ya que estaríamos haciendo una mala gestión de la memoria de nuestra aplicación.

Para ello podemos utilizar el decorador @property. De esta forma, el método se ejecutará cada vez que lo llamemos y se generará la información sin tener que almacenarla en memoria. Vamos a verlo con el siguiente ejemplo.

```python
from dataclasses import dataclass
from datetime import date, timedelta

def daterange(start_date: date, end_date: date):
    for n in range(int((end_date - start_date).days)):
        yield start_date + timedelta(n)

@dataclass
class PeriodoTemporal:
  fecha_inicio: date
  fecha_final: date

  @property
  def fechas(self):
    return [_date for _date in daterange(start_date=self.fecha_inicio, end_date=self.fecha_final)]
```

Cada vez que llamamos a la propiedad fechas, nos devuelve mediante list comprehension todas las fechas incluidas entre fecha_inicio y fecha_final.
https://www.w3schools.com/python/python_lists_comprehension.asp

```python
periodo1 = PeriodoTemporal(
    fecha_inicio = date(2022, 10, 1),
    fecha_final=date(2022, 10, 20)
)

periodo1.fechas
```

Cuando hacemos 2 > 1 obtenemos True, pero ¿porqué?. Básicamente la clase int tiene definidos unos métodos que permiten comparar dos objetos del mismo tipo y poder obtener un orden. ¿Podríamos hacer lo mismo con cualquier otra clase? Sí!. Además seremos nosotros los encargados de definir el criterio para poder compararlos.

Vamos a coger el ejemplo anterior y pensar la lógica antes de implementarla. ¿Cómo definimos que un Periodo es mayor que otro? Fijándonos en el atributo fecha_inicial sería una buena aproximación. Si un período tiene una fecha_inicial mayor que otro período diremos que el primero es mayor. Una fecha es mayor que otra si es posterior.

En este caso vamos a acudir a otro decorador conocido como @total_ordering. Este va a permitir que definiendo únicamente un par de métodos seamos capaces de aplicar todos los operadores de comparación como >, < e ==.

```python
from __future__ import annotations #Esto permite definir un atributo del mismo tipo que de la clase donde se encuentra.

from functools import total_ordering
from dataclasses import dataclass
from datetime import date, timedelta

@dataclass
@total_ordering
class Periodo:
  fecha_inicio: date
  fecha_fin: date

  def __eq__(self, otro_periodo: Periodo):
    return (self.fecha_inicio, self.fecha_fin) == (self.otro_periodo.fecha_inicio, self.otro_periodo.fecha_fin)

  def __lt__(self, otro_periodo: Periodo):
    return self.fecha_inicio < otro_periodo.fecha_inicio
```

Sólo hemos definido dos métodos:

1. __eq__: aquí indicamos el criterio para que dos períodos sean considerados iguales, en este caso que tengan las mimas fechas iniciales y finales.  
2. __lt__: Aquí indicamos el criterio para que el objeto de la instancia sea considerado menor que otro. En este caso, que su fecha inicial sea menor que la del otro objeto.

Vamos a ver cómo funciona


```python
periodo1 = Periodo(
    fecha_inicio=date(2022, 10, 1),
    fecha_fin=date(2022, 10, 15),
)

periodo2 = Periodo(
    fecha_inicio=date(2022, 12, 1),
    fecha_fin=date(2022, 12, 30),
)
```

```python
periodo1 < periodo2
```

Para poder diferenciar de forma única un objeto y poderlos utilizar en extructuras como sets o como claves de diccionarios es imprescindible implementar el método built-in __hash__

Si probamos a crear un set a partir de los objetos periodo1 y periodo2, obtendremos un error.

```python
#unique_set_of_periods = {periodo1, periodo2}
```

```python
from __future__ import annotations #Esto permite definir un atributo del mismo tipo que de la clase donde se encuentra.

from functools import total_ordering
from dataclasses import dataclass
from datetime import date, timedelta

@dataclass
@total_ordering
class Periodo:
  fecha_inicio: date
  fecha_fin: date

  def __eq__(self, otro_periodo: Periodo):
    return (self.fecha_inicio, self.fecha_fin) == (self.otro_periodo.fecha_inicio, self.otro_periodo.fecha_fin)

  def __lt__(self, otro_periodo: Periodo):
    return self.fecha_inicio < otro_periodo.fecha_inicio

  def __hash__(self):
    return hash((self.fecha_inicio, self.fecha_fin))

periodo1 = Periodo(
    fecha_inicio=date(2022, 10, 1),
    fecha_fin=date(2022, 10, 15),
)

periodo2 = Periodo(
    fecha_inicio=date(2022, 12, 1),
    fecha_fin=date(2022, 12, 30),
)
```

```python
unique_set_of_periods = {periodo1, periodo2}
dict_of_periods = {
    periodo1: "primer_periodo",
    periodo2: "segundo_periodo",
}
```

El uso de @dataclass permite implementar estos métodos de forma automática únicamente indicando frozen y eq igual a True, veámoslo.

```python
from __future__ import annotations #Esto permite definir un atributo del mismo tipo que de la clase donde se encuentra.

from functools import total_ordering
from dataclasses import dataclass
from datetime import date, timedelta

@dataclass(frozen=True, eq=True)
@total_ordering
class Periodo:
  fecha_inicio: date
  fecha_fin: date

  def __lt__(self, otro_periodo: Periodo):
    return self.fecha_inicio < otro_periodo.fecha_inicio

periodo1 = Periodo(
    fecha_inicio=date(2022, 10, 1),
    fecha_fin=date(2022, 10, 15),
)

periodo2 = Periodo(
    fecha_inicio=date(2022, 12, 1),
    fecha_fin=date(2022, 12, 30),
)

unique_set_of_periods = {periodo1, periodo2}
dict_of_periods = {
    periodo1: "primer_periodo",
    periodo2: "segundo_periodo",
}
```

Por último y para cerrar la herencia en Python vamos a ver una última aproximación que nos permite forzar la implementación de métodos cuando heredamos de un objeto.

Para ello, creamos nuestro objeto padre, y cada uno de los métodos que queremos que sean implementados por las clases que hereden, añadiremos el decorador @abstractmethod.

Veamos un ejemplo.

```python
from abc import ABC, abstractmethod

class Poligono(ABC):
  @abstractmethod
  def numero_de_lados(self):
    ...

@dataclass
class TrianguloEquilatero(Poligono):
  lado: int

  @property
  def area(self):
    print((self.lado**2)/2)
```

```python
triangulo1 = TrianguloEquilatero(lado = 10)
triangulo1.area
```

Nos fuerza a implementar el método número de lados si queremos que TrianguloEquilatero herede de Poligono.

```python
from abc import ABC, abstractmethod

class Poligono(ABC):
  @abstractmethod
  def numero_de_lados(self):
    ...

@dataclass
class Hexagono(Poligono):
  lado: int

  @property
  def area(self):
    print((self.lado**2)/2)

  def numero_de_lados(self):
    print("Tres lados")

  def numero_de_lados_Sergio(self):
    print("Seis lados")

triangulo1 = TrianguloEquilatero(lado = 10)
triangulo1.area
triangulo1.numero_de_lados()
```

# Decoradores

Todo en Python es un objeto. Desde los métodos hasta las clases. Esto implica que podemos pasar una función a otra función como atributo.

Ese es el concepto de los decoradores: una forma simplificada que permite pasar un método como atributo a otro método y modificar su comportamiento.

Veamos el siguiente ejemplo:

```python
def funcion_decoradora(func):
  def metodo_decorador(*args, **kwargs):
    print("Antes de que se ejecute el método")
    result = func(*args, **kwargs)
    print("El método ya se ha ejecutado")
    return result
  return metodo_decorador

def imprime_mi_nombre(nombre: str):
  print(nombre)
```

Tenemos el método llamado funcion_decoradora, la cual recibe una función como atributo. Dentro de ésta, tenemos la función metodo_decorador que muestra por pantalla antes y después de que se ejecute la función pasada como argumento.

Vamos a ver como se usa.

```python
imprime_mi_nombre_modificado = funcion_decoradora(imprime_mi_nombre)
```

¿Qué hemos hecho? Al pasar a la funcion_decoradora la función imprime_mi_nombre, funcion_decoradora nos ha devuelto una versión modificada de imprime_nombre, ya que nos ha devuelto el metodo_decorador.

```python
imprime_mi_nombre_modificado(nombre='Sergio')
```

Una forma simplificada de decorar cualquier función es la siguiente:

```python
def funcion_decoradora(func):
  def metodo_decorador(*args, **kwargs):
    print("Antes de que se ejecute el método")
    func(*args, **kwargs)
    print("El método ya se ha ejecutado")
  return metodo_decorador

@funcion_decoradora
def imprime_mi_nombre(nombre: str):
  print(nombre)
```

```python
imprime_mi_nombre(nombre='Sergio')
```

Ahora la función imprime_mi_nombre ya tiene su comportamiento modificado indicando el decorador al comienzo de la definición de la función.

# Excepciones

Podemos crear nuestras propias excepciones para cuando falle parte del código seamos más rápidos a la hora de localizar el error. Nuestras implementaciones deberán heredar de la clase Exception.

```python
class CarConstructorException(Exception):
  pass

class Car:
  def __init__(
      self,
      modelo: str,
      potencia: int
  ):

    self.modelo = modelo
    if potencia < 0:
      raise CarConstructorException
    self.potencia = potencia

try:
  coche1 = Car(
      modelo= 'Seat',
      potencia= -110,
  )

except CarConstructorException:
  print("Potencia erronea")
```

## Resumen de decoradores:

Resumen de los decoradores vistos hasta el momento:

*   @staticmethod: Nos permite asociar un método a la clase y no a la instancia.
*   @property: Permite definir atributos que se calculan on demand.
*   @dataclass: Crear clases sencillas sin necesidad de definir un constructor.
*   @total_ordering: Para ordenar objetos de cualquier clase sin tener que definir todos los métodos necesarios.
