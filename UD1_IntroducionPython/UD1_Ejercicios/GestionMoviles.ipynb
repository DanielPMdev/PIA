Enunciado
Crea una clase llamada Movil que represente un teléfono móvil con los siguientes atributos:

marca (cadena)

modelo (cadena)

almacenamiento (entero, en GB)

bateria (entero, porcentaje entre 0 y 100, valor inicial 100)

uso_total (entero, en horas, valor inicial 0)

Métodos que debe tener la clase
especificaciones()
Muestra toda la información del móvil (marca, modelo, almacenamiento, batería y uso total).

usar(horas)

Aumenta el uso total en la cantidad indicada de horas.

Disminuye la batería en un 10 % por cada hora de uso.

Si la batería llega a 0 o menos, muestra un mensaje de error y no permite seguir usándolo.

cargar()
Restaura la batería al 100 %.

nivel_bateria()
Devuelve o muestra el nivel actual de la batería.

```python
class Movil:
  def __init__(self, marca, modelo, almacenamiento, bateria=100, uso_total=0):
    self.marca = marca
    self.modelo = modelo
    self.almacenamiento = almacenamiento
    self.bateria = bateria
    self.uso_total = uso_total

  def especificaciones(self):
    print(f"Marca: {self.marca}")
    print(f"Modelo: {self.modelo}")
    print(f"Almacenamiento: {self.almacenamiento} GB")
    print(f"Batería: {self.bateria}%")
    print(f"Uso total: {self.uso_total} horas")

  def usar(self, horas):
    self.horas = horas

    if self.bateria > 0:
      self.uso_total += horas
      self.bateria -= horas * 10
      print(f"Se ha usado el móvil durante {self.horas} horas.")
      print(f"Tiene un uso total de {self.uso_total} horas.")
      print(f"Le queda al dispositivo un {self.bateria} % de batería.")
    else:
      print("Error: La batería se ha agotado no se puede usar")

  def cargar(self):
    self.bateria = 100
    return print(f"Se ha cargado la batería al {self.bateria} %")

  def nivel_bateria(self):
    return print(f"El dispositivo tiene un {self.bateria} % de batería")
```

```python
miMovil = Movil("Xiaomi", "Readmi Note 10 Pro", 128)
```

```python
miMovil.especificaciones()
```

```python
miMovil.usar(8)
```

```python
miMovil.nivel_bateria()
```

```python
miMovil.cargar()
```

```python
miMovil.nivel_bateria()
```

```python
miMovil.usar(5)
```
