### 4. Aplicar las operaciones aritméticas básicas sobre objetos Series.

Operación,Operador,Método de Series

*   Suma,+,".add(other, fill_value=None)"
*   Resta,-,".sub(other, fill_value=None)"
*   Multiplicación,*,".mul(other, fill_value=None)"
*   División,/,".div(other, fill_value=None)":









```python
serie_a = pd.Series([10, 20, 30])
serie_b = pd.Series([5, 10, 15])

serie_suma = serie_a.add(serie_b)
serie_resta = serie_a.sub(serie_b)
serie_multiplicacion = serie_a.mul(serie_b)
serie_division = serie_a.div(serie_b)
```

```python
print(serie_suma)
print("-------------")
print(serie_resta)
print("-------------")
print(serie_multiplicacion)
print("-------------")
print(serie_division)
```

### 5. Usar operadores relacionales para comparar objetos Series. (==, >, <)

Operador,Método de Series,Descripción

*   "==,.eq(),Igual a"
*   "!=,.ne(),No igual a"
*   ">,.gt(),Mayor que"
*   "<,.lt(),Menor que"
*   ">=,.ge(),Mayor o igual que"
*   <=,.le(),Menor o igual que






```python
serie_ventas = pd.Series([300, 200, 300], index=['Ene', 'Feb', 'Mar'])
serie_meta = pd.Series([150, 200, 450], index=['Ene', 'Feb', 'Mar'])
```

```python
print(serie_ventas.eq(serie_meta))
print("-------------")
print(serie_ventas.ne(serie_meta))
print("-------------")
print(serie_ventas.gt(serie_meta))
print("-------------")
print(serie_ventas.lt(serie_meta))
print("-------------")
print(serie_ventas.ge(serie_meta))
print("-------------")
print(serie_ventas.le(serie_meta))
```

### 6. Convertir un diccionario Python en un objeto Series.

```python
diccionario_poblacion = {
    'España': 47.6,
    'Francia': 67.8,
    'Alemania': 83.2,
    'Italia': 59.1
}

print("--- Diccionario Python ---")
print(type(diccionario_poblacion))
print(diccionario_poblacion)


# 2. Convertir el diccionario a Series
series_poblacion = pd.Series(diccionario_poblacion)

print("\n--- Objeto Pandas Series ---")
print(type(series_poblacion))
print(series_poblacion)
```

### 7. Convertir un array NumPy en un objeto Series.

```python
array_temperaturas = np.array([25, 28, 22, 26, 20])

print("--- Array NumPy ---")
print(type(array_temperaturas))
print(array_temperaturas)


# 2. Convertir el array a Series
# Por defecto, Pandas asignará un índice entero (0, 1, 2, ...)
series_temperaturas = pd.Series(array_temperaturas)

print("\n--- Objeto Pandas Series Resultante (Índice por defecto) ---")
print(type(series_temperaturas))
print(series_temperaturas)

# 3. Conversión especificando un Índice
# También puedes proporcionar etiquetas personalizadas para el índice.
dias = ['Lun', 'Mar', 'Mié', 'Jue', 'Vie']
series_con_indice = pd.Series(array_temperaturas, index=dias)

print("\n--- Objeto Pandas Series (Índice Personalizado) ---")
print(series_con_indice)
```

### 8. Cambiar el tipo de datos de un objeto Series. to_numeric

### 18. Obtener los elementos pares e impares de un objeto Series numérico.

```python
s = pd.Series([1, 2, 3, 4, 5, 6, 7, 8])

pares = s[s % 2 == 0]
impares = s[s % 2 != 0]

print("Pares:\n", pares)
print("Impares:\n", impares)
```

### 20. Calcular el valor absoluto de los elementos de un objeto Series.

### 21. Reemplazar por un valor arbitrario los valores de una serie que no satisfagan una condición.

```python
s = pd.Series([10, 20, 30, 40, 50])

# Conservar valores > 25, el resto reemplazarlos por -1
# .where mantiene los que cumplen la condición, reemplaza los que NO
s_modificada = s.where(s > 25, -1)

print(s_modificada)
```

## Ejercicios DataSet

```python
import pandas as pd
import numpy as np

nombre = ['Oliva', 'Daniela', 'Juan', 'Germán', 'Edward', 'Alex', 'Julio',
          'Edgar', 'Angie', 'Irlesa']
puntaje = [11.5, 8, 15.5, np.nan, 8, 19, 13.5, np.nan, 8, 18]
intentos = [1, 3, 2, 3, 2, 3, 1, 1, 2, 1]
califico = ['Sí', 'No', 'Sí', 'No', 'No', 'Sí', 'Sí', 'No', 'No', 'Sí']

indices = ['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h', 'i', 'j']

jugadores = {'nombre': nombre, 'puntaje': puntaje, 'intentos': intentos,
             'califico': califico}

df = pd.DataFrame(data=jugadores, index=indices)

df
```

### 16. Ordenar en modo descendente un DataFrame por diferentes columnas.




```python
# Ordena por 'puntaje' (descendente: de mayor a menor) y luego por 'intentos'
# (ascendente: de menor a mayor) como criterio de desempate.
ordenado = df.sort_values(by=['puntaje', 'intentos'], ascending=[False, True])
print(ordenado)
```

### 17. Usar la función mask para modificar los valores que cumplan una condición.




```python
# Reemplazar con 0 los puntajes menores a 10
# Mask reemplaza donde la condición es VERDADERA
df_masked = df.mask(df['puntaje'] < 10, 0)
print(df_masked)
```

### 18. Realizar consultas sobre los elementos de un objeto DataFrame con la función query.




```python
# Consultar jugadores con más de 1 intento y puntaje mayor a 10
resultado = df.query('intentos > 1 and puntaje > 10')
print(resultado)
```
