```python
# 6. ❌

# Con [0, :] mostramos la primera fila
# Con [:, 0] mostramos la primera columna

array6 = np.arange(20).reshape(5, 4)

array6
```

```python

for i in range(4):
    print(array6[:, i].max())
    print(array6[:, i].min())
    print(array6[i, :].max())
    print(array6[i, :].min())

print(array6.max())
print(array6.min())

print(array6.max(axis=0))
print(array6.min(axis=0))
print(np.max(array6))

# np.max(matriz) muestra el valor maximo de esa matriz 
# y matriz.max(axis=0) muestra la fila donde se encuentra ese valor


```
