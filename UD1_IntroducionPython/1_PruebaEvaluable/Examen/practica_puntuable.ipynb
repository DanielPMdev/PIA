### Ejercicio 6

6. (1 pt) Crea una matriz de 5×4 con los valores que desees (pueden ser enteros). Muestra por pantalla:
- Los valores máximos para cada columna
- Los valores mínimos para cada columna
- Los valores máximos para cada fila
- Los valores mínimos para cada fila
- El valor máximo global
- El valor mínimo global
- La posición (índices) del máximo
- La posición (índices) del mínimo
Además, explica qué diferencia hay entre usar np.max(matriz) y matriz.max(axis=0).

```python
matriz = np.random.randint(1, 50, size=(5, 4))

print("Matriz 5x4:\n")
print(matriz)
print("\n" + "="*50)

# Valores máximos para cada columna (Eje 0)
max_cols = matriz.max(axis=0)
# Valores mínimos para cada columna (Eje 0)
min_cols = matriz.min(axis=0)

# Valores máximos para cada fila (Eje 1)
max_filas = matriz.max(axis=1)
# Valores mínimos para cada fila (Eje 1)
min_filas = matriz.min(axis=1)

print(f"Máximos por columna (axis=0): {max_cols}")
print(f"Mínimos por columna (axis=0): {min_cols}")
print(f"Máximos por fila (axis=1):   {max_filas}")
print(f"Mínimos por fila (axis=1):   {min_filas}")
print("="*50)
```

```python
# Valor máximo global
max_global = np.max(matriz)
# Valor mínimo global
min_global = np.min(matriz)

# Posición (índices) del máximo global
pos_max = np.unravel_index(np.argmax(matriz), matriz.shape)
# Posición (índices) del mínimo global
pos_min = np.unravel_index(np.argmin(matriz), matriz.shape)


print(f"Valor máximo global: {max_global}")
print(f"Valor mínimo global: {min_global}")
print(f"Posición del máximo (fila, col): {pos_max}")
print(f"Posición del mínimo (fila, col): {pos_min}")
```

### Ejercicio 7


7. (1 pt) Escribir un programa que pregunte al usuario por las ventas de un rango de años y muestre por pantalla una Serie con los datos de las ventas indexada por los años, antes y después de aplicarles un descuento del 10%. (Nota: con que preguntes dos años es suficiente).

```python
inicio = int(input("Introduce el año inicial: "))
fin = int(input("Introduce el año final: "))


datos_ventas = {}
for año in range(inicio, fin + 1):
  venta = float(input(f"Introduce las ventas del año {año}: "))
  if venta <= 0:
    print("Cantidad no válida. Se asignará 0.")
    datos_ventas[año] = 0.0
  else:
    datos_ventas[año] = venta
```

```python
ventas_serie = pd.Series(datos_ventas)
ventas_serie.name = "Ventas"
ventas_serie.index.name = "Año"
ventas_serie
```

```python
descuento_serie = ventas_serie * 0.9
descuento_serie.name = "Ventas con Descuento"
descuento_serie.index.name = "Año"
descuento_serie
```

a) Indica el tipo del índice de la serie
resultante.

```python
print(f"\n a) Tipo del índice de la serie: {ventas_serie.index.dtype}")
```

b) Muestra el año con ventas máximas y el año con ventas mínimas antes y después del descuento.

```python
print("\n b) Años con máximos y mínimos:")
print(f"   - Año máx ventas (original): {ventas_serie.idxmax()} (Valor: {ventas_serie.max()})")
print(f"   - Año mín ventas (original): {ventas_serie.idxmin()} (Valor: {ventas_serie.min()})")

#Al aplicar un % a todas las ventas, siempre serán los mismos
print(f"   - Año máx ventas (descuento): {descuento_serie.idxmax()} (Valor: {descuento_serie.max()})")
print(f"   - Año mín ventas (descuento): {descuento_serie.idxmin()} (Valor: {descuento_serie.min()})")
```

### Ejercicio 11


11. (opcional) Con el DataFrame titanic (ya sin edades nulas):

a) Representa un histograma de edades diferenciando por sexo (dos histogramas superpuestos o en subplots).


```python
edades_hombres = df_sin_nulos_edad[df_sin_nulos_edad['Sex'] == 'male']['Age']
edades_mujeres = df_sin_nulos_edad[df_sin_nulos_edad['Sex'] == 'female']['Age']
```

```python
plt.hist(edades_hombres, bins=20, alpha=0.5, label='Hombres', color='blue', edgecolor='black')
plt.hist(edades_mujeres, bins=20, alpha=0.5, label='Mujeres', color='orange', edgecolor='black')

plt.title('Distribución de Edades por Sexo')
plt.xlabel('Edad')
plt.ylabel('Frecuencia')
plt.legend()
plt.grid(axis='y', alpha=0.3)
```


b) Representa un gráfico de barras con el porcentaje de supervivientes por clase.


```python
plt.subplot(1, 2, 2) # 1 fila, 2 columnas, gráfico 2

# Calculamos el porcentaje de supervivencia agrupando por clase
# La media de una columna de ceros (0) y unos (1) nos da la proporción. Multiplicamos por 100.
supervivencia_clase = df_sin_nulos_edad.groupby('Pclass')['Survived'].mean() * 100

# Creamos el gráfico de barras
colores = ['green', 'yellow', 'red'] # Colores distintivos para 1ª, 2ª y 3ª clase
plt.bar(supervivencia_clase.index, supervivencia_clase.values, color=colores, edgecolor='black')

plt.title('Porcentaje de Supervivencia por Clase')
plt.xlabel('Clase (Pclass)')
plt.ylabel('Porcentaje de Supervivencia (%)')
plt.xticks([1, 2, 3], ['1ª Clase', '2ª Clase', '3ª Clase']) # Etiquetas claras en el eje X
plt.ylim(0, 100) # Eje Y de 0 a 100%
plt.grid(axis='y', alpha=0.3)

# Mostrar etiquetas de valor encima de las barras (Opcional pero recomendado)
for i, v in zip(supervivencia_clase.index, supervivencia_clase.values):
    plt.text(i, v + 1, f"{v:.1f}%", ha='center', fontweight='bold')

# Ajustar el diseño y mostrar
plt.tight_layout()
plt.show()
```
