Para practicar herencia y decoradores, quiero que desarrolléis lo siguiente:
Vamos a crear la clase Animal con los siguientes atributos:

1.   especie: str
2.   nombre: str
3.   edad: int
4.   nacimiento: date

Dentro de la clase Animal vais a definir los siguientes métodos:

1.   imprime_animal: permitirá imprimir la información del animal: especie y nombre
2.   meses: Propiedad que devolverá todos los años que ha vivido el animal. Por ejemplo: Si tiene 2 años y nació el 22-10-2020, obtendremos [2020, 2021, 2022].

Además de la clase Animal deberéis definir dos clases que hereden de Animal, como Mamifero, Oviparo. Estas clases tendran atributos extra como en el caso del mamifero el número de meses de gestación o en el caso de los ovíparos el número de días que tarda en eclosionar. Sin embargo continuarán disponiendo de los atributos de la clase padre.

Debemos ser capaces de ordenar los animales en función de su fecha de nacimiento por lo que deberemos definir los métodos que nos permitan poder comparar objetos de la misma clase.

A continuación crea una clase Conversor que permita mediante métodos estáticos transformar metros en pies y pulgadas a centímetros.

Finalmente, si te ves con ganas, crea un decorador que permita calcular el tiempo de ejecución de un método, una pista:



```python
from datetime import date
from functools import total_ordering

@total_ordering
class Animal():
    def __init__(self, especie: str, nombre: str, edad: int, nacimiento: date):
        self.especie = especie
        self.nombre = nombre
        self.edad = edad
        self.nacimiento = nacimiento

    def imprime_animal(self):
        print(f"Especie: {self.especie}, Nombre: {self.nombre}")

    def meses(self):
        start_year = self.nacimiento.year
        end_year = self.nacimiento.year + self.edad
        return list(range(start_year, end_year + 1))

    def __eq__(self, other):
        if not isinstance(other, Animal):
            return NotImplemented
        return self.nacimiento == other.nacimiento

    def __lt__(self, other):
        if not isinstance(other, Animal):
            return NotImplemented
        return self.nacimiento < other.nacimiento
```

```python
class Mamifero(Animal):
    def __init__(self, especie: str, nombre: str, edad: int, nacimiento: date, meses_gestacion: int):
        super().__init__(especie, nombre, edad, nacimiento)
        self.meses_gestacion = meses_gestacion

class Oviparo(Animal):
  def __init__(self, especie, nombre, edad, nacimiento, num_dias_eclosionar):
      super().__init__(especie, nombre, edad, nacimiento)
      self.num_dias_eclosionar = num_dias_eclosionar
```

```python
class Conversor:
    @staticmethod
    def metros_a_pies(metros):
        """Convierte metros a pies."""
        return metros * 3.28084

    @staticmethod
    def pulgadas_a_centimetros(pulgadas):
        """Convierte pulgadas a centímetros."""
        return pulgadas * 2.54
```

```python
import time

def medir_tiempo(func):
    """Decorador que mide el tiempo de ejecución de una función."""
    def wrapper(*args, **kwargs):
        inicio = time.time()
        resultado = func(*args, **kwargs)
        fin = time.time()
        print(f"⏱ Tiempo de ejecución de {func.__name__}: {fin - inicio:.6f} segundos")
        return resultado
    return wrapper
```

```python
# ======== 1️⃣ Pruebas con Animal, Mamifero y Oviparo ========

# Crear algunos animales
perro1 = Animal("Perro", "Rex", 5, date(2019, 4, 12))
gato = Animal("Gato", "Miau", 3, date(2021, 6, 5))
perro2 = Animal("Perro", "Rex", 5, date(2019, 4, 12))  # mismo nacimiento que perro1

# Mostrar información
perro1.imprime_animal()   # Especie: Perro, Nombre: Rex
gato.imprime_animal()   # Especie: Gato, Nombre: Miau

# Comparaciones
print("¿perro1 == perro2?", perro1 == perro2)     # True, mismo nacimiento
print("¿perro1 < gato?", perro1 < gato)       # True, perro1 es mayor
print("¿gato > perro1?", gato > perro1)       # True

# Método personalizado
print("Años vividos por perro1:", perro1.meses())

# Crear subclases
mamifero = Mamifero("Elefante", "Dumbo", 10, date(2015, 2, 1), meses_gestacion=22)
print(f"{mamifero.especie} tiene {mamifero.meses_gestacion} meses de gestación.")

oviparo = Oviparo("Tortuga", "Shelly", 2, date(2023, 3, 10), num_dias_eclosionar=90)
print(f"{oviparo.especie} tarda {oviparo.num_dias_eclosionar} días en eclosionar.")


# ======== 2️⃣ Pruebas con la clase Conversor ========

print("\n--- Pruebas de conversión ---")
print("10 metros son:", Conversor.metros_a_pies(10), "pies")
print("25 pulgadas son:", Conversor.pulgadas_a_centimetros(25), "cm")


# ======== 3️⃣ Pruebas con el decorador medir_tiempo ========

@medir_tiempo
def tarea_pesada():
    """Simula una tarea lenta."""
    time.sleep(1.5)
    return "Tarea completada"

resultado = tarea_pesada()
print(resultado)


# ======== 4️⃣ Decorador aplicado a un método estático ========

class ConversorConTiempo:
    @staticmethod
    @medir_tiempo
    def metros_a_pies(metros):
        time.sleep(0.5)
        return metros * 3.28084

print("\n--- Conversor con tiempo ---")
print("5 metros son:", ConversorConTiempo.metros_a_pies(5), "pies")
```
