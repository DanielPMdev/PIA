```python
# Estimar el precio de una vivienda

# Imagina que el precio son 50.000€ y por cada dormitorio son 50.000€
# 1 casa de 2 dormitorios = 150.000€
```

```python
import tensorflow as tf
import numpy as np
from tensorflow import keras
```

```python
# Datos para el modelo
# Declaramos los vectores de entrenamiento : y = 2x -1
```

```python
def precio_vivienda():
  # Define los vectores de entrenamiento
  x = np.array([2, 5, 1, 2, 3, 4], dtype=float)
  y = np.array([15, 30, 10, 15, 20, 25], dtype=float)

  # Define la red neuronal. Una simple de una simple neurona.
  model = tf.keras.Sequential([keras.layers.Dense(units=1, input_shape=[1])])

  # Compila y entrena la red neuronal
  model.compile(optimizer='sgd', loss='mean_squared_error')

  # Entrenamos el modelo
  model.fit(x, y, epochs=500)

  return model
```

```python
precio_vivienda()
```

```python
new_home = np.array([7.0], dtype=float)
prediction = precio_vivienda().predict(new_home)
```
