```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, SimpleRNN, LSTM, GRU, Dense
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
import numpy as np
```

```python
#Datos de ejemplo

texts = [
    "me encanta este dГ­a",
    "odio que llueva",
    "soy maravillosa",
    "no me gusta madrugar",
    "no me gusta nada",
    "es increible"
]
# 1 es positivo, 0 es negativo
labels = [1,0,1,0,0,1]


```

```python
#Tokenizar

tokenizer = Tokenizer()
tokenizer.fit_on_texts(texts)
sequences = tokenizer.texts_to_sequences(texts)
```

```python
X = pad_sequences(sequences)
y = np.array(labels)

vocab_size= len(tokenizer.word_index) + 1
max_length =X.shape[1]
```

```python
print("vocabulario", tokenizer.word_index)
print("tamaГ±o", vocab_size)
print("long", max_length)
```

```python
#Modelo RNN

model_rnn = Sequential([
    Embedding(input_dim=vocab_size, output_dim=16, input_length=max_length),
    SimpleRNN(16),
    Dense(1, activation="sigmoid")
  ])

model_rnn.compile(optimizer="adam", loss="binary_crossentropy", metrics=["accuracy"])

model_rnn.fit(X,y, epochs=15, verbose=1)
```
