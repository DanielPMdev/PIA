```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from keras.models import Sequential
from keras.layers import Dense
from sklearn.preprocessing import RobustScaler

import numpy as np
```

```python

# Cargar los datos
df = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/PIA 23-24 ok/diabetes.csv')
```

```python
print(df.info())
print(df.describe())
```

```python
# variables objetivo
X = df.drop('Outcome', axis=1)
y = df['Outcome']

```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python

scaler = RobustScaler()

X_train_prep = scaler.fit_transform(X_train)

X_test_prep = scaler.transform(X_test)
```

```python
model = Sequential([
    Dense(16, activation='relu', input_shape=(8,)),  # 8 características
    Dense(16, activation='relu'),
    Dense(1, activation='sigmoid')  # Salida binaria
])
```

```python
print("Forma del conjunto de entrenamiento:", X_train.shape)
print("Forma del conjunto de prueba:", X_test.shape)
```

```python
model.compile(optimizer='adam', loss='binary_crossentropy')
```

```python
history = model.fit(X_train, y_train, epochs=100, batch_size=10, validation_split=0.2)
```

```python
loss = model.evaluate(X_test, y_test)
print(f"Loss: {loss}")
```

```python
predicciones = model.predict(X_test)
```

```python
# Realizar predicciones
predicted_classes = (predicciones > 0.5).astype(int).flatten()
```

```python
# Asumiendo que 'predicted_classes' y 'y_test' están definidos
fallo =0
# Iterar a través de las predicciones y las etiquetas reales
for i in range(len(predicted_classes)):
    # Comprobar si la predicción es diferente de la etiqueta real
    if predicted_classes[i] != y_test.iloc[i]:
        fallo += 1
        print(f"Índice: {i}, Predicción: {predicted_classes[i]}, Real: {y_test.iloc[i]}")

print(" Ha fallado en ", fallo, "de un total de", len(predicted_classes))
```

```python
aciertos=100-((100*fallo)/len(predicted_classes))
print(f"Porcentaje de acierto: {np.round(aciertos, 2)}%")
```
